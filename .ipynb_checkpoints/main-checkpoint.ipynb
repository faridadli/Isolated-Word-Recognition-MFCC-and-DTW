```python
# Getting the input + visualization #
```

```python
import librosa as lr
import matplotlib.pyplot as plt
import librosa.display as lrd
TEMPLATE_FILE = "assets/hello_there.wav"
TARGET_FILE = "assets/obi_wan_hello_there.wav"
SAMPLING_RATE = 44000
FIG_DIMENSION = (14, 7)
```

```python
template_data, template_sr = lr.load(TEMPLATE_FILE, sr=SAMPLING_RATE)
target_data, target_sr = lr.load(TARGET_FILE, sr=SAMPLING_RATE)
```

```python
plt.figure(figsize=FIG_DIMENSION)
lrd.waveplot(template_data, sr=SAMPLING_RATE)
```

```python
plt.figure(figsize=FIG_DIMENSION)
lrd.waveplot(target_data, sr=SAMPLING_RATE)
```

```python
# Data pre-processing #
```

```python
import librosa.util as lru
import preprocess
THRESHOLD = 0.003
```

```python
template_data = lr.effects.preemphasis(template_data)
target_data = lr.effects.preemphasis(target_data)

template_data = preprocess.noise_gating(template_data, (-THRESHOLD, THRESHOLD))
target_data = preprocess.noise_gating(target_data, (-THRESHOLD, THRESHOLD))

preprocessed_template = preprocess.zero_align(template_data)
preprocessed_target = preprocess.zero_align(target_data)
preprocessed_target = preprocess.zero_align(preprocessed_target)
```

```python
plt.figure(figsize=FIG_DIMENSION)
lrd.waveplot(preprocessed_template, sr=SAMPLING_RATE)
```

```python
plt.figure(figsize=FIG_DIMENSION)
lrd.waveplot(preprocessed_target, sr=SAMPLING_RATE)
```

```python
# Feature extraction #
```

```python
from librosa.feature import mfcc
```

```python
template_data = mfcc(y = template_data, sr=SAMPLING_RATE)
target_data = mfcc(y = target_data, sr=SAMPLING_RATE)
```

```python
# Feature matching #
```

```python
from fastdtw import fastdtw
```

```python
distance, path = fastdtw(template_data, target_data)
distance
```
